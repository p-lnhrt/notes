```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
```

```python
df = pd.DataFrame(data={'registration': (['F-HPJA']*6+['F-HPJB']*6)*2,
                        'time': [1,2,3,9,10,11,3,4,5,10,11,12]*2,
                        'feature': ['featureX']*12+['featureY']*12,
                        'feature_value': [1,4,-1,4,2,3,6,1,3,5,3,2,3,6,2,3,5,3,2,3,6,4,2,3], 
                        'key': ['a', 'a', 'a', 'b', 'b', 'b', 'c', 'c', 'c', 'd', 'd', 'd',
                               'e', 'e', 'e', 'f', 'f', 'f', 'g', 'g', 'g', 'h', 'h', 'h']},
                 columns={'registration', 'time', 'feature', 'feature_value', 'key'})
```

```python
def clustered_lineplot(x, y, *args, **kwargs):
    # on pourrait y ajouter la génération des clés à l'intérieur
    # la fonction prendrait un argument additionnel timedelta
    ax = plt.gca()
    data = kwargs.pop('data') # dataframe de la facet(i,j) passé dans le kwarg 'data' par map_dataframe
    color = kwargs.pop('color') # couleur de la facet(i,j,hue) passée dans le kwarg 'color' par map_dataframe
    _keys = data.key.unique()
    
    for key in _keys:
        ax.plot(x, y, data=data[data.key==key], color=color)
        # ax.plot ne prend les dimensions 'x' et 'y' qu'en positionnel
```

```python
def build_legend_data(facet_grid):
    # Retourne un dictionnaire dont:
    # - les clés sont les labels de chaque élément de la légende
    # - les valeurs sont des artists destiné à servir de vignette pour chaque élément de la légende
    # Ces objets destinés à devenir des vignettes sont souvent appelés '(legend) handles'.
    # Ici nos vignettes/handles seront des rectangles.
    current_palette = sns.color_palette()
    hues = facet_grid.hue_names
    
    legend_data = {}
    for i, level_name in enumerate(hues):
        artist handle = mpatches.Patch(color=current_palette[i], label=level_name) 
        legend_data.update({level_name: artist handle})
        
    return legend_data
```

```python
cfg = sns.FacetGrid(df, col='registration', hue='feature')\
    .map_dataframe(clustered_lineplot, x='time', y='feature_value')

# vu qu'il existe plusieurs objets Line2D pour la même couleur, seaborn/matplotlib ne semble pas vouloir choisir
# on est donc obligé de construire notre propre légende qu'on passe à la méthode add_legend
legend_data = build_legend_data(cfg)
cfg.add_legend(legend_data=legend_data, title='custom_legend');
```
